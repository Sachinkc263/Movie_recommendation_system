# Data Preprocessing

This notebook performs data preprocessing on **The Movies Dataset** (Kaggle - Rounak Banik) to merge datasets, create features, and prepare data for modeling.

## Preprocessing Tasks:
- Merge datasets (movies, credits, keywords, links)
- Process ratings and create user-item matrix
- Feature normalization
- Train/test split
- Create user and movie features

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Import from src modules
import sys
sys.path.append('..')
from src.data import DataLoader, DataPreprocessor

# Initialize data loader and preprocessor
data_loader = DataLoader()
preprocessor = DataPreprocessor()

print("Modules loaded successfully!")
```

## Load Cleaned Data

```python
# Load cleaned data
movies_metadata = data_loader.load_movies_metadata(clean=True)
credits = data_loader.load_credits(clean=True)
keywords = data_loader.load_keywords(clean=True)
links = data_loader.load_links(clean=True)
ratings = data_loader.load_ratings(clean=True)

print(f"Movies Metadata: {movies_metadata.shape}")
print(f"Credits: {credits.shape}")
print(f"Keywords: {keywords.shape}")
print(f"Links: {links.shape}")
print(f"Ratings: {ratings.shape}")
```

## Merge Datasets

```python
# Merge datasets
merged_data = preprocessor.merge_datasets(
    movies_clean=movies_metadata,
    credits_clean=credits,
    keywords_clean=keywords,
    links_clean=links
)

print(f"Merged data shape: {merged_data.shape}")
print("\nFirst few rows:")
merged_data.head()
```

## Process Ratings

```python
# Process ratings
ratings_processed = preprocessor.process_ratings(ratings)

print(f"Processed ratings shape: {ratings_processed.shape}")
print("\nFirst few rows:")
ratings_processed.head()
```

## Create User-Item Matrix

```python
# Create user-item matrix
user_item_matrix, user_mapping, movie_mapping = preprocessor.create_user_item_matrix(ratings_processed)

print(f"User-item matrix shape: {user_item_matrix.shape}")
print(f"Number of users: {len(user_mapping)}")
print(f"Number of movies: {len(movie_mapping)}")
```

## Normalize Features

```python
# Normalize numerical features
numerical_features = ['budget', 'revenue', 'runtime', 'popularity', 'vote_average', 'vote_count']
merged_normalized = preprocessor.normalize_features(merged_data, numerical_features)

print(f"Normalized data shape: {merged_normalized.shape}")
print("\nFirst few rows:")
merged_normalized.head()
```

## Create Genre Dummies

```python
# Create genre dummy variables
merged_with_genres, top_genres = preprocessor.create_genre_dummies(merged_normalized)

print(f"Data with genre dummies shape: {merged_with_genres.shape}")
print("\nGenre columns:")
genre_cols = [col for col in merged_with_genres.columns if col.startswith('genre_')]
print(genre_cols[:10])
```

## Train/Test Split

```python
# FIX #1 — User-based split (replaces temporal split)
# Each user's ratings are split 80/20 randomly so EVERY test user
# has training history — eliminating the coverage gap.
train_ratings, test_ratings = preprocessor.create_train_test_split(
    ratings_processed,
    test_size=0.2,
    method='user_based',   # was 'temporal' — this is the root-cause fix
    min_ratings_per_user=5,
    random_state=42,
)

# Overlap enforcement (redundant after user_based split but kept as a safety check)
overlap_users = set(train_ratings['userId']) & set(test_ratings['userId'])
train_ratings = train_ratings[train_ratings['userId'].isin(overlap_users)]
test_ratings  = test_ratings[test_ratings['userId'].isin(overlap_users)]

print(f"Train ratings shape: {train_ratings.shape}")
print(f"Test ratings shape:  {test_ratings.shape}")
print(f"Overlap users:       {len(overlap_users):,}")
print(f"Train ratio:         {len(train_ratings)/( len(train_ratings)+len(test_ratings) ):.2f}")

# VERIFICATION GATE: all evaluated models need ≥ 95 users
assert len(overlap_users) >= 95, f"Overlap too small: {len(overlap_users)}"
print("✓ Verification gate passed: overlap users ≥ 95")
```

## Create User Features

```python
# Create user features
user_features = preprocessor.create_user_features(train_ratings)

print(f"User features shape: {user_features.shape}")
print("\nFirst few rows:")
user_features.head()
```

## Create Movie Features

```python
# Create movie features
movie_features = preprocessor.create_movie_features(train_ratings)

print(f"Movie features shape: {movie_features.shape}")
print("\nFirst few rows:")
movie_features.head()
```

## Save Preprocessed Data

```python
# Save preprocessed data
preprocessor.save_preprocessed_data(
    movies_integrated=merged_with_genres,
    train_ratings=train_ratings,
    test_ratings=test_ratings,
    user_features=user_features,
    movie_features=movie_features,
    user_item_matrix=user_item_matrix,
    user_to_index=user_mapping,
    movie_to_index=movie_mapping
)

print("Preprocessed data saved successfully!")
```

## Preprocessing Summary

```python
# Create summary
summary_data = {
    'Dataset': ['Merged Data', 'Train Ratings', 'Test Ratings', 'User Features', 'Movie Features'],
    'Rows': [len(merged_with_genres), len(train_ratings), len(test_ratings), len(user_features), len(movie_features)],
    'Columns': [len(merged_with_genres.columns), len(train_ratings.columns), len(test_ratings.columns), len(user_features.columns), len(movie_features.columns)]
}

summary_df = pd.DataFrame(summary_data)
print("Data Preprocessing Summary:")
print(summary_df.to_string(index=False))
```
